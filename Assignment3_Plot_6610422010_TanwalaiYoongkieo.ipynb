## 3. Plot
- จากข้อมูลใน part2 ในส่วนนี้เลยจะมาทำ plot เพิ่มเติมเพื่อหาว่า Female กลุ่มใดที่ platform นี้ควรทำการเจาะตลาดเพิ่มเติม
- โดยจะแบ่งกลุ่ม Female ออกตามช่วงอายุ

```python
def age_segment(x):
    if x > 60:
        return 'over 60'
    elif x >= 50:
        return '50-60'
    elif x >= 40:
        return '40-49'
    elif x >= 30:
        return '30-39'
    elif x >= 20:
        return '20-29'
    else :
        return 'under 20'
```

```python
df['age_segment'] = df['Age'].apply(age_segment)
```

```python
df.head()
```

```python
avg_amount = df[df['Gender'] == 'Female'].groupby(['Category','age_segment'])['Purchase Amount (USD)'].mean().reset_index(name = 'avg_amount')
```

```python
num_customer = df[df['Gender'] == 'Female'].groupby(['Category','age_segment'])['Customer ID'].count().reset_index(name = 'total_customer')
```

```python
df_plot = pd.merge(avg_amount,num_customer,left_on=['Category','age_segment'],right_on=['Category','age_segment'])
```

```python
df_plot.head()
```

```python
scatter = px.scatter(
    df_plot,
    x = 'total_customer',
    y = 'avg_amount',
    color = 'age_segment',
    facet_col = 'Category'
)
scatter.show()
```
